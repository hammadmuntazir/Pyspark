###Requirement:
####1. Answer the below question using the dev.spark_db.sf_fire_calls
What are top 3 zip codes that accounted for most calls?


```python
%sql
SELECT CallType,Zipcode,count(*) as count 
FROM  dev.spark_db.sf_fire_calls
Where CallType is NOT NULL
GROUP BY CallType,Zipcode
Order BY count DESC
LIMIT 3
```

#### 2. Answer the above question using the Spark DataFrame API


2.1 Create a DataFrame reading data from the table


```python
# Read data and Create DataFrame
fire_df=spark.read.table("dev.spark_db.sf_fire_calls")
```

2.2 Apply Transformations


```python
df_1=fire_df.select('CallType','Zipcode')
df_2=df_1.where("CallType is NOT NULL")
df_3=df_2.groupBy('CallType','Zipcode').count()
df_4=df_3.orderBy("count",ascending=False)
df_5=df_4.limit(3)
```

```python
# Apply an action method
df_5.show()
```
